```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('/Users/jacob_kitz/Desktop/hgn_processing/HGN_Classification/testing/mr_kitty.jpeg')
```

```python
# scaling
res = cv.resize(img, None, fx=2, fy=2, interpolation=cv.INTER_CUBIC) # double size

cv.imshow('Mr. Kitty', res)

cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
# translation
rows, cols, _ = img.shape

M = np.float32([[1, 0, 100], [0, 1, 50]]) # translation matrix
dst = cv.warpAffine(img, M, (cols, rows)) # apply translation

cv.imshow('Mr. Kitty', dst)

cv.waitKey(1) 
cv.destroyAllWindows()
```

```python
# rotation

M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),90,1)
dst = cv.warpAffine(img,M,(cols,rows))

cv.imshow('Mr. Kitty', dst)

cv.waitKey(1) 
cv.destroyAllWindows()
```

```python
# showing with plt now bc the windows get annoying
# mildly more advanced rotation

rows,cols,ch = img.shape
 
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
 
M = cv.getAffineTransform(pts1,pts2)
 
dst = cv.warpAffine(img,M,(cols,rows))
 
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
# perspective change
# you can't really see it as much with this picture but its better in the documentation

rows,cols,ch = img.shape
 
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
 
M = cv.getPerspectiveTransform(pts1,pts2)
 
dst = cv.warpPerspective(img,M,(300,300))
 
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
